The shape of the distribution varies based on α and β:

    If α = β = 1, it's uniform (flat line)
    If α > 1 and β > 1, it's bell-shaped
    If α < 1 and β < 1, it's U-shaped
    If α > 1 and β < 1, it peaks near 1
    If α < 1 and β > 1, it peaks near 0

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = 2
beta = 4
x = np.random.beta(alpha, beta, 10000)
plt.hist(x, bins=100)
```

```python
mean = alpha / (alpha + beta)
mean
```

```python
np.mean(x)
```

```python
y = np.linspace(0, 1, 100)
k = 10
alpha = y * k
beta = (1 - y) * k
plt.plot(y, alpha, label='alpha')
plt.plot(y, beta, label='beta')
plt.legend()
```

```python
i = 10
yi = y[i]
ai = alpha[i]
bi = beta[i]

x = np.random.beta(ai, bi, 10000)
mean_dist = ai / (ai + bi)
mean_observed = np.mean(x)
std = np.std(x)

plt.hist(x, bins=100)
plt.axvline(mean_dist, color='red', label=f'mean_dist = {mean_dist:.2f}')
plt.axvline(mean_observed, color='green', label=f'mean_observed = {mean_observed:.2f}')
plt.title(f"y = {yi:.2f}, alpha = {ai:.2f}, beta = {bi:.2f}, std = {std:.2f}")
plt.legend()
plt.show()
```
