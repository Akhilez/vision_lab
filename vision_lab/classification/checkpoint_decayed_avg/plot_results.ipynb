```python
plt.figure(figsize=(10, 6))

for decay_rate in df.decay_rate.unique():
    df_sub = df[df["decay_rate"] == decay_rate]
    plt.plot(df_sub.epoch, df_sub.val_accuracy, label=f"Decay rate = {decay_rate:0.1f}", marker="o")

plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Decaying average of model parameters")
plt.legend()
```
